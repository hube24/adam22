```python
dictionary = {}
mus = ""
```

```python

transcripts = glob.glob("transcripts/*")
```

```python

for file in transcripts:
    with open(file) as data:
        current_name = ""
        current_seq = ""

        for line in data:
            if line[0] == ">":
                
                if current_name != "":
                    dictionary[current_name] = current_seq
                current_name = line[1:].strip()
                current_seq = ""

                if "intron" in line:
                    break
            else:
                current_seq+=line.strip()

        
            
```

```python
with open("adam22.txt") as data:
    for line in data:
        mus += line.strip()
```

```python
MARK_TAG_OPEN = "<mark>"
MARK_TAG_CLOSED = "</mark>"
```

```python
mus_html = mus
```

```python

count = 1
for key in dictionary:
    index = mus_html.find(dictionary[key])
    print(dictionary[key], index)
    elen = len(dictionary[key])
    if(index > 0):
        mus_html = mus_html[:(index+elen)] + "<b>(Ex." + str(count) + "-" + str(elen) + "bp)</b>" + MARK_TAG_CLOSED + mus_html[(index+elen):]
        mus_html = mus_html[:(index)] + "<mark id='" + key + "'>" +  "<b>(Ex." + str(count) + "-" + str(elen) + "bp)</b>" + mus_html[(index):]
        count+=1
        
```

```python
f = open("site1.html", "w")
f.write(mus_html.strip())
f.close()
```
