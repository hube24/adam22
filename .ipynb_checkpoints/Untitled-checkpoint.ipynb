```python
dictionary = {}
dictionary_not_found = {}

mus = ""
```

```python

transcripts = glob.glob("transcripts/*")
```

```python

for file in transcripts:
    with open(file) as data:
        current_name = ""
        current_seq = ""

        for line in data:
            if line[0] == ">":
                
                if current_name != "":
                    dictionary[current_name] = current_seq
                current_name = line[1:].strip()
                current_seq = ""

                if "intron" in line:
                    break
            else:
                current_seq+=line.strip()

        
            
```

```python
with open("adam22.txt") as data:
    for line in data:
        mus += line.strip()
```

```python
MARK_TAG_OPEN = "<mark>"
MARK_TAG_CLOSED = "</mark>"
```

```python
mus_html = mus
index_list = []
dictionary_not_found = {}
id_dict = {}
# index, key, type, count, len
count = 1
for key in dictionary:
    index = mus.find(dictionary[key])
    elen = len(dictionary[key])
    if(index > 0):
        index_list.append((index, "S", elen, key))        
        index_list.append((index+elen, "E", elen, key)) 
        count+=1
    else:
        dictionary_not_found[key] = dictionary[key]
        
index_list.sort()         

shift_right = 0

l_index, l_itype, l_elen, l_key = -1,"",-1,""
current_id = 0
forbidden_keys = []


for (index, itype, elen, key) in index_list:
    
    
        if itype == "E" and key not in forbidden_keys:
            mus_html = mus_html[:(index+shift_right)] + MARK_TAG_CLOSED + mus_html[(index+shift_right):]
            shift_right += len(MARK_TAG_CLOSED)
        
        if itype == "S":
            # check if last exists, check if last len and start == current len and start,
            #     if yes then give the same id and change type of end element to none 
            #     if no then increase id and write mark
            
            if l_index != -1:
                if l_index == index and l_elen == elen and l_itype == itype:
                    print("duplicate")
                    id_dict[key] = current_id
                    forbidden_keys.append(key)
                else:
                    print("new")
                    current_id+=1
                    id_dict[key] = current_id
                    mus_html = mus_html[:(index+shift_right)] + "<mark id='" + str(id_dict[key]) + "'>" + mus_html[(index+shift_right):]
                    shift_right += (len("<mark id='") + len(str(id_dict[key])) + len("'>"))
            else:
                id_dict[key] = current_id
                mus_html = mus_html[:(index+shift_right)] + "<mark id='" + str(id_dict[key]) + "'>" + mus_html[(index+shift_right):]
                shift_right += (len("<mark id='") + len(str(id_dict[key])) + len("'>"))

        l_index, l_itype, l_elen, l_key = index, itype, elen, key
```

```python
f = open("site1.html", "w")
f.write(mus_html.strip())
f.close()
```

```python
count=1
list_html = ""
list_not_found = ""
for key in dictionary: 
    if not key in dictionary_not_found.keys():
        list_html += "<li> <a href='#" + str(id_dict[key]) + "'>" + "<b>Ex." + str(count) + "</b> " + key + "</a></li>" + "\n"
        count+=1

for key in dictionary_not_found: 
    list_not_found += "<li><a><b>Ex." + key + "</b> | " + dictionary_not_found[key] + "</a></li>" + "\n"
    
f = open("site2.html", "w")
f.write(list_html)
f.close()

f = open("site3.html", "w")
f.write(list_not_found)
f.close()
```
